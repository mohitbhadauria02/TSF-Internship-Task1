To Predict the percentage of marks of the students based on the number of hours they studied

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

There is no null value in the Dataset so, we can now visualize our Data.

```python
df.plot(x='Hours', y='Scores',markerfacecolor='blue', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
X = df[['Hours']]
y = df[['Scores']] 
```

From the above scatter plot there looks to be correlation between the 'Marks Percentage' and 'Hours Studied', Lets plot a regression line to confirm the correlation.



```python
sns.regplot(x= df['Hours'], y= df['Scores'])
plt.title('Regression Plot',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
print(df.corr())
```

It is confirmed that the variables are positively correlated.

### Training the Model


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

### Fitting the Data into the model

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
print("Training complete.")
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

Evaluating the Model


```python
from sklearn import metrics
print('Mean absoulute Error:',
     metrics.mean_absolute_error(y_test,y_pred))
```

## What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?

```python
hours = 9.25
np_hour = np.array(hours)
pred = lr.predict(np_hour.reshape(1,-1))
print('No. of hours = 9.25')
print('Predicted score =',pred[0])
```

According to the regression model if a student studies for 9.25 hours a day he/she is likely to score 93.69 marks.
